```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
#split input and output
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
#split train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={'kernel':['rbf','poly','sigmoid','linear'],
            'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(), param_grid, refit=True, verbose=3,n_jobs=-1)
# fitting the model for grid search
grid.fit(x_train, y_train)
```

```python
#print best parameter after tuning
print(grid.best_params_)
```

```python
#(cv_results_) contains a dictionary of results from the cross-validation search.
re=grid.cv_results_
print("The R_score value for best parameter is {}:".format(grid.best_params_))
```

```python
#convert table
table=pd.DataFrame.from_dict(re)
table
```

```python
grid_pred=grid.predict(x_test)
grid_pred
```

```python
#validation
from sklearn.metrics import r2_score
r2_score=r2_score(grid_pred,y_test)
r2_score
```

```python
#save the model
import pickle
filename="final_svmr_model.sav"
pickle.dump(grid,open(filename,"wb"))
loaded_model=pickle.load(open("final_svmr_model.sav","rb"))
loaded_model
```
