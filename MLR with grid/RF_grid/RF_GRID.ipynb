```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'criterion':['mse','mae'],'max_features':['auto','sqrt','log2'],'n_estimators':[10,100]}
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)

#fitting the model for grid search
grid.fit(independent,dependent)
```

```python
#print best parameter after tuning
#print (grid.best_params_)
re=grid.cv_results_
#print(re)
grid_pred=grid.predict(x_test)
#grid_pred
#print("The R_score value for best parameter{}:".format(grid.best_params_))
```

```python
from sklearn.metrics import r2_score
r2_score=r2_score(grid_pred,y_test)
r2_score
```

```python
table=pd.DataFrame.from_dict(re)
table
#print(type(table))
```

```python
excel_tab=table.to_excel('output.xlsx', index=False)
excel_tab
```

```python
print("R_score value for best parameter{}:".format(grid.best_params_),r2_score)
```
